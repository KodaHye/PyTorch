#### 파이토치를 활동한 딥러닝 구현의 흐름

1. 전처리, 후처리, 네트워크 모델의 입출력 확인
2. 데이터셋 작성
   * **Dataset 클래스** 생성
     * 입력 데이터와 라벨 등을 쌍으로 갖는 클래스
     * 데이터에 대한 전처리 클래스의 인스턴스를 할당하여 파일을 읽을 때 자동으로 전처리를 적용하도록 설정,,,,?
     * 훈련 데이터, 검증 데이터, 테스트 데이터에 대한 Dataset 작성
3. 데이터 로더 작성
   * **DataLoader 클래스** 생성
     * 데이터를 어떻게 가져올지 설정하는 클래스
4. **네트워크 모델** 작성
5.  네트워크 모델의 **순전파함수** 정의
6.  **손실함수** 정의
   * **오차 역전파법**을 하기 위한 손실함수 정의
     * **오차 역전파법**을 통해 **결합 파라미터의 오차경사**를 구할 수 있음
7. **최적화 기법** 설정
   * 네트워크 모델의 결합 파라미터를 학습시킬 때의 최적화 기법
8. 학습/검증 실시
9. 테스트 데이터로 추론
